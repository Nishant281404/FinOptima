```python
class CryptoTradingEnv(gym.Env):
    def __init__(self, df):
        super(CryptoTradingEnv, self).__init__()

        self.df = df
        self.current_step = 0
        self.done = False

        # Define the action space (buy, sell, hold)
        self.action_space = gym.spaces.Discrete(3)  # 3 actions: 0 - hold, 1 - buy, 2 - sell

        # Define the observation space (the feature space of the dataset)
        self.observation_space = gym.spaces.Box(low=0, high=1, shape=(df.shape[1],), dtype=np.float32)

        # Initialize state
        self.state = None

    def reset(self):
        self.current_step = 0
        self.done = False
        self.state = self.df.iloc[self.current_step].values
        return self.state

    def step(self, action):
        self.current_step += 1

        if self.current_step >= len(self.df) - 1:
            self.done = True
            return self.state, 0, self.done, {}

        # Get the current and next states
        prev_state = self.state
        self.state = self.df.iloc[self.current_step].values

        # Calculate reward (difference between predicted close price and actual close price)
        predicted_close = self.state[1]  # Assuming predicted price is from the model (second column)
        actual_close = self.df.iloc[self.current_step]['Close']
        
        reward = self.get_reward(predicted_close, actual_close)

        # Implement the trading logic (buy, sell, hold)
        if action == 1:  # Buy action
            # Assume the agent buys at the predicted price
            reward -= 0.01  # Small penalty for transaction
        elif action == 2:  # Sell action
            # Assume the agent sells at the predicted price
            reward += 0.01  # Small profit for transaction
        
        return self.state, reward, self.done, {}

    def get_reward(self, predicted_close, actual_close):
        # Reward function based on prediction accuracy
        # Reward: +1 for correct prediction, -1 for wrong prediction
        return 1 - abs(predicted_close - actual_close)  # Simple reward function
```

```python
class QLearningAgent:
    def __init__(self, action_space, state_space, learning_rate=0.1, discount_factor=0.99, epsilon=0.1):
        self.action_space = action_space
        self.state_space = state_space
        self.learning_rate = learning_rate
        self.discount_factor = discount_factor
        self.epsilon = epsilon

        # Initialize Q-table (state-action values)
        self.q_table = np.zeros((state_space, action_space))

    def choose_action(self, state):
        if np.random.rand() < self.epsilon:
            return np.random.choice(self.action_space)  # Random action (exploration)
        else:
            return np.argmax(self.q_table[state])  # Best action (exploitation)

    def learn(self, state, action, reward, next_state):
        predict = self.q_table[state, action]
        target = reward + self.discount_factor * np.max(self.q_table[next_state])  # Bellman equation
        self.q_table[state, action] = predict + self.learning_rate * (target - predict)
```

```python
class QLearningAgent:
    def __init__(self, action_space, state_space, learning_rate=0.1, discount_factor=0.99, epsilon=0.1, bins=5):
        self.action_space = action_space
        self.state_space = state_space
        self.learning_rate = learning_rate
        self.discount_factor = discount_factor
        self.epsilon = epsilon
        self.bins = bins

        # Initialize Q-table with the shape (bins, ..., bins, action_space)
        # We create bins for each state feature
        self.q_table = np.zeros([self.bins] * self.state_space + [self.action_space])

    def discretize_state(self, state):
        # Convert continuous state to discrete values (bin index)
        state_bins = []
        for feature in state:
            state_bins.append(np.digitize(feature, np.linspace(min(feature), max(feature), self.bins)))
        return tuple(state_bins)

    def choose_action(self, state):
        # Discretize the state to get the index for the Q-table
        state_discretized = self.discretize_state(state)
        if np.random.rand() < self.epsilon:
            return np.random.choice(self.action_space)  # Exploration
        else:
            return np.argmax(self.q_table[state_discretized])  # Exploitation

    def learn(self, state, action, reward, next_state):
        # Discretize the states
        state_discretized = self.discretize_state(state)
        next_state_discretized = self.discretize_state(next_state)

        # Apply Q-learning update rule
        predict = self.q_table[state_discretized + (action,)]
        target = reward + self.discount_factor * np.max(self.q_table[next_state_discretized])
        self.q_table[state_discretized + (action,)] = predict + self.learning_rate * (target - predict)
```

```python
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim

# Q-learning agent class
class QLearningAgent:
    def __init__(self, action_space, state_space, learning_rate=0.1, discount_factor=0.99, epsilon=0.1, bins=5):
        self.action_space = action_space
        self.state_space = state_space
        self.learning_rate = learning_rate
        self.discount_factor = discount_factor
        self.epsilon = epsilon
        self.bins = bins

        # Initialize Q-table with the shape (bins, ..., bins, action_space)
        self.q_table = np.zeros([self.bins] * self.state_space + [self.action_space])

    def discretize_state(self, state):
        # Convert continuous state to discrete values (bin index)
        state_bins = []
        for feature in state:
            state_bins.append(np.digitize(feature, np.linspace(min(feature), max(feature), self.bins)))
        return tuple(state_bins)

    def choose_action(self, state):
        # Discretize the state to get the index for the Q-table
        state_discretized = self.discretize_state(state)
        if np.random.rand() < self.epsilon:
            return np.random.choice(self.action_space)  # Exploration
        else:
            return np.argmax(self.q_table[state_discretized])  # Exploitation

    def learn(self, state, action, reward, next_state):
        # Discretize the states
        state_discretized = self.discretize_state(state)
        next_state_discretized = self.discretize_state(next_state)

        # Apply Q-learning update rule
        predict = self.q_table[state_discretized + (action,)]
        target = reward + self.discount_factor * np.max(self.q_table[next_state_discretized])
        self.q_table[state_discretized + (action,)] = predict + self.learning_rate * (target - predict)

# Deep Q-learning agent class
class DQNAgent:
    def __init__(self, state_size, action_size, learning_rate=0.001, discount_factor=0.99):
        self.state_size = state_size
        self.action_size = action_size
        self.learning_rate = learning_rate
        self.discount_factor = discount_factor

        # Create Q-network
        self.model = self.build_model()

    def build_model(self):
        # Neural network for Q-learning
        model = nn.Sequential(
            nn.Linear(self.state_size, 64),
            nn.ReLU(),
            nn.Linear(64, 64),
            nn.ReLU(),
            nn.Linear(64, self.action_size)
        )
        return model

    def choose_action(self, state):
        state = torch.FloatTensor(state).unsqueeze(0)
        q_values = self.model(state)
        action = torch.argmax(q_values).item()
        return action

    def learn(self, state, action, reward, next_state):
        state = torch.FloatTensor(state).unsqueeze(0)
        next_state = torch.FloatTensor(next_state).unsqueeze(0)

        q_values = self.model(state)
        next_q_values = self.model(next_state)

        # Q-learning target
        target = reward + self.discount_factor * torch.max(next_q_values)

        # Loss function (Mean Squared Error)
        loss = nn.MSELoss()(q_values[0][action], target)

        # Optimize the model
        optimizer = optim.Adam(self.model.parameters(), lr=self.learning_rate)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
```
